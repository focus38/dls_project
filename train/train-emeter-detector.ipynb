```python
import os
import numpy as np
import pandas as pd
from ultralytics import YOLO
```

```python
model = YOLO('yolo11n.pt')
data = '/kaggle/input/emeter3/data.yaml'
project_dir = '/kaggle/working/runs/train'
# Обучение модели
# Описание параметров для настройки модели https://docs.ultralytics.com/ru/usage/cfg/#predict-settings
results = model.train(data=data, epochs=250, batch=16, imgsz=640, device='0', workers=4, optimizer='AdamW',
    verbose=True, save=True, save_period=10, project=project_dir, name='exp', exist_ok=False,
    patience=50, max_det=50, half=False, plots=True)
```
